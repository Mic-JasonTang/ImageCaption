```python
import requests
import json
url = "http://0.0.0.0:5000/imagecaption/do"
try:
    result = requests.post(url)
    result = json.loads(result.content.decode('utf-8'))
    if result['status'] == 'ok':
        print(result)
    else:
    print("网络错误，请稍后再试")
except Exception as e:
    print(e)
```
